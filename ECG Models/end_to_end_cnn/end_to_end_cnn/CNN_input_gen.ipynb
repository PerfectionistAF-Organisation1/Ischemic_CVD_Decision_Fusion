```python
import wfdb as wfdb
import matplotlib 
```

```python
record = wfdb.rdrecord('100/100', sampto=3000)
```

```python
print(record)
```

```python
annotation = wfdb.rdann('100/100', 'atr', sampto=3000)
```

```python
print(annotation)
```

```python
wfdb.plot_wfdb(record=record, annotation=annotation, plot_sym=True,
                   time_units='seconds', title='MIT-BIH Record 100 Annotated',
                   figsize=(10,4), ecg_grids='all')
```
